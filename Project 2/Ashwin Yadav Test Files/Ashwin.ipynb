```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales.csv")
```

```python
plt.figure()
plt.plot(df['sales_year'], df['running_total'], color='aqua')
plt.xlabel("year")
plt.ylabel("running total sales")
plt.title("running total of sales by year")
plt.show()
```

## 5

```python
df = pd.read_csv("territory.csv")
```

```python
plt.figure()
plt.bar(df['territory_name'], df['total_sales'], color='aqua')
plt.xticks(rotation=45)
plt.xlabel("territory")
plt.ylabel("total sales")
plt.title("territory wise ranking")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("customer order.csv")
```

```python
plt.figure()
plt.hist(df['average_order_value'], bins=20, color='aqua')
plt.xlabel("average order ")
plt.ylabel("number of customers")
plt.title("distribution of average order")
plt.show()
```
